### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
# Advice: Implement input validation across the subplots to ensure robustness.
import numpy as np
import matplotlib.pyplot as plt

# Sample data for demonstration
x = np.arange(1, 6)
y = np.array([2, 4, 6, 8, 10])
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 45, 10]

# Input validation
def validate_line_data(x, y):
    if not (isinstance(x, np.ndarray) and isinstance(y, np.ndarray)):
        raise ValueError("x and y must be numpy arrays.")
    if x.shape != y.shape:
        raise ValueError("x and y must have the same shape.")

def validate_bar_data(categories, values):
    if not (isinstance(categories, list) and isinstance(values, list)):
        raise ValueError("categories and values must be lists.")
    if len(categories) != len(values):
        raise ValueError("categories and values must have the same length.")

def validate_scatter_data(x, y):
    validate_line_data(x, y)

def validate_pie_data(values):
    if not isinstance(values, list):
        raise ValueError("values must be a list.")
    if not all(isinstance(v, (int, float)) for v in values):
        raise ValueError("All pie values must be numeric.")
    if sum(values) <= 0:
        raise ValueError("Sum of pie values must be positive.")

# Validate data
validate_line_data(x, y)
validate_bar_data(categories, values)
validate_scatter_data(x, y)
validate_pie_data(values)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart
axs[0, 0].plot(x, y, marker='o')
axs[0, 0].set_title('Line Chart')

# Bar chart
axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(x, y, color='r')
axs[1, 0].set_title('Scatter Plot')

# Pie chart
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')

plt.tight_layout()
plt.show()
```
